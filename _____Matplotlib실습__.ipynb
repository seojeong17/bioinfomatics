### **1) 기본 그래프**



plot()함수를 이용해 선 그래프를 그려보도록 하겠습니다.
plot은 선 (Line) 또는 마커 (Marker) 그래프 그리기에 적합

image.png

````python
plt.plot([1,2,3,4])   #값을 한 개만 넣으면 y축 먼저 지정
#plt.show()
````

plt.show() 시각화 명령은 실제 차트로 렌더링하고 마우스 움직임등의 이벤트를 기다리라는 지시
주피터노트북이나 코랩등은 셀단위로 플롯명령을 자동 렌더링 해주므로 show() 명령 필요 없음

* plot() 함수는 리스트의 값들이 y값이라고 가정하고 x값을 자동으로 만들어냅니다. 따라서 리스트 [1, 2, 3, 4] 를 y값이라고 가정하고 0, 1, 2, 3을 x값으로 자동으로 부여합니다.

* x값과 y값을 모두 입력할 수도 있습니다. x값과 y값이 모두 입력되면, 해당 점들을 이은 꺾은선 그래프가 그려집니다.

* plot() 함수의 데이터는 **리스트** 뿐만 아니라 **튜플**, **nparray** 형태로도 입력할 수 있습니다.

* show() 함수는 그래프가 화면에 나타나게 하는 함수입니다.

````python
plt.plot([1,2,3,4],[1,4,9,16])    # 리스트 형태
plt.show()                        # 값이 두 개 들어가면 첫번째 x축, 두번째 y축
````

````python
plt.plot((1,2,5,10))      #튜플
plt.show()
````

````python
plt.plot(np.array([2,3,5,10]))    #np.array 형태
plt.show()
````

````python
plt.plot([1,2,3,4],[1,4,9,16],'c^')
plt.axis([0,6,0,20])
plt.show
````

x, y 값 인자에 대해 선의 색상과 형태를 지정하는 포맷 문자열 (Format string)을 세번째 인자에 입력할 수 있습니다.

포맷 문자열 ‘ro’는 빨간색 (‘red’)의 원형 (‘o’) 마커를 의미합니다.

또한, 예를 들어 ‘b-‘는 파란색 (‘blue’)의 실선 (‘-‘)을 의미합니다.

axis() 함수를 이용해서 축의 범위 [xmin, xmax, ymin, ymax]를 지정

image.png

plot() 함수에 **딕셔너리** 데이터를 입력해 그래프를 그릴 수도 있습니다.

````python
data_dict = {'x':[1,2,3,4],'y':[1,3,7,13],}
plt.plot('x','y', data = data_dict)
plt.show()
````


* 'x'와 'y'를 key, x값과 y값을 value로 지정한 딕셔너리형 데이터를 형성해 data_dict에 저장했습니다.

* plot() 함수에 key값과 데이터를 입력하면, x값에 [1,2,3,4], y값에 [1,3,7,13]를 입력한 그래프와 동일한 그래프가 그려집니다.

plot() 함수로 **여러 개의 곡선**을 하나의 그래프로 그릴 수 있습니다.
<br><br>
numpy의 arange 함수를 이용해 0부터 5까지 0.2간격인 데이터(0, 0.2, 0.4, 0.6, ..., 5)를 형성해 여러 개의 곡선 그래프 생성

````python
t = np.arange(0.,5.,0.2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
````

* (t, t, 'r--'), (t, t\*\*2, 'bs'), (t, t**3, 'g^')가 3개의 곡선 각각의 인자에 해당합니다. 즉, 한 곡선에 해당하는 인자를 연속으로 입력하면, 여러 곡선이 한 그래프에 나타나게 됩니다.

* 이때, 각 곡선의 세번째 인자는 선의 색상과 형태를 지정합니다. 　　***'r--' (빨간 대쉬), 'bs' (파란 사각형), 'g^' (녹색 삼각형)***

### **2) 축 레이블 설정하기**



xlabel(), ylabel() 함수를 사용하면 x축, y축에 대한 레이블을 표시할 수 있습니다.

````python
plt.plot([1,2,3,4],[1,4,9,16])
plt.xlabel('age',labelpad=15,loc = 'right')   #lablepad는 여백
plt.ylabel('height', labelpad=15, loc = 'bottom')
plt.show()
````


* xlabel(), ylabel() 함수를 이용해 x축의 레이블은 'x_label', y축의 레이블은 'y_label'로 설정했습니다.
* **labelpad=15**

  labelpad 파라미터는 축 레이블의 여백(padding)을 지정합니다.

  xlabel() 함수의 loc 파라미터는 X축 레이블의 위치를 지정합니다. ({‘left’, ‘center’, ‘right’})

 ylabel() 함수의 loc 파라미터는 Y축 레이블의 위치를 지정합니다. ({‘bottom’, ‘center’, ‘top’})


### **3) 범례 표시하기**



legend() 함수를 사용하면 그래프에 범례를 표시할 수 있습니다. plot() 함수에서 label 문자열을 지정하고, legend() 함수를 호출하면 그래프에 범례가 표시됩니다.

````python
plt.plot([1,2,3,4],[1,4,9,16],label = 'price')
plt.legend (loc = 'best')      #loc = location
plt.show()
````



* **loc='lower right**

  loc 파라미터는 범례의 위치를 설정합니다.

  * 'best' : 범례의 최적의 위치 / 'center' : 중앙
  * 'lower left' : 왼쪽 하단 / 'lower right' : 오른쪽 하단
  * 'upper left' : 왼쪽 상단 / 'upper right' : 오른쪽 상단
  * 'center left' : 왼쪽 중앙 / 'center left' : 오른쪽 중앙

### **4) 축 범위 지정하기**



xlim(), ylim() 함수를 사용하면 x축, y축의 범위를 지정할 수 있습니다.

````python
plt.plot([1,2,3,4],[1,3,7,12])
plt.xlim((0,5))
plt.ylim([0,16])
plt.show()
````



* xlim() 함수에 xmin, xmax 값을 리스트 또는 튜플의 형태로 입력하면 x축의 범위가 xmin부터 xmax까지로 지정됩니다.

* ylim() 함수에 ymin, ymax 값을 리스트 또는 튜플의 형태로 입력하면 y축의 범위가 ymin부터 ymax까지로 지정됩니다.

* 입력값이 없으면 데이터에 맞게 자동으로 범위를 지정합니다.

ㅤ

axis() 함수를 사용해서 x축, y축의 범위를 지정할 수도 있습니다.

* **plt.axis([0, 5, 0, 16])**

  axis( [xmin, xmax, ymin, ymax] ) 함수에 xmin, xmax, ymin, ymax 값을 입력하면 x축의 범위를 xmin부터 xmax까지, y축의 범위를 ymin부터 ymax까지로 범위를 지정할 수 있습니다.

### **2) Bar Plot**

- 값을 막대의 높이로 나타내므로 여러 항목의 수량이 많고 적음을 한눈에 알아보기 쉬워 여러 항목 데이터를 비교할 대 주로 사용

```
plt.bar(x, height [,width, color, tick_label, align, label])
```
- x: 막대의 x 좌표
- height: 막대로 표시할 데이터
- width: 막대의 폭을 지정(0~1사이의 실수를 지정, 기본값 0.8)
- color:막대그래프의 색깔을 지정
- tick_label:막대그래프 각각의 이름을 지정(기본값: 숫자)
- align: x위치를 기준으로 막대그래프의 정렬(기본값:center)
- Label: 범례에 표시할 이름을 지정

- 제주도 지역의 2010년 ~ 2012년 방문한 내국인과 외국인 통계자료

````python
x = list(range(3))
year = ['2010','2011','2012']
domestic = [6801, 7695, 8010]
foreign = [777, 1046, 1684]

plt.bar(x, domestic)
plt.bar(x, foreign)
plt.xticks(x, year)
plt.show()
````

````python
x1 = list(range(3))
x2 = [i+0.3 for i in x1]    # 간격x +3, 원래 크기의 25%
year = ['2010','2011','2012']
domestic = [6801, 7695, 8010]
foreign = [777, 1046, 1684]

plt.bar(x1, domestic, width=0.25)
plt.bar(x2, foreign, width = 0.25)
plt.xticks(x, year)
plt.show()
````

````python
x1 = list(range(3))
x2 = [i+0.3 for i in x1]    # 간격x +3, 원래 크기의 25%
year = ['2010','2011','2012']
domestic = [6801, 7695, 8010]
foreign = [777, 1046, 1684]

plt.barh(x1, domestic, height = 0.25 , color='r')
plt.barh(x2, foreign, height = 0.25, color='g')
plt.yticks(x, year)
plt.show()
````

- 히스토그램은 hist() 명령을 사용하고, bins 인수로 데이터를 집계할 구간 정보를 받으며 반환 값으로 데이터 집계 결과를 반환

````python
import random

heights = [random.randint(150,180)for _ in range(30)]
print(heights)
````

````python
plt.hist(heights, bins =13, cumulative= True)   # cumulative 누적값
plt.hist(heights, bins =13, cumulative= False)
plt.xlabel('height')
plt.ylabel('frequency')
plt.show()
````

`plt.hist(x,bins=None, range=None, density=False, weights=None, cumulative=False, bottom=None, histtype='bar', align='mid', orientation='vertical', rwidth=None, log=False, color=None, label=None, stacked=False, **kwargs)`



*   x : 데이터값으로 리스트나 ndarray 타입으로 넣음
*   bins : 막대(bins)의 갯수
*   range : bin의 범위
*   density : Normalize를 함.
*   weights : x에 있는 모든 값의 가중치(y값의미) , x와 모양 같아야함
*   cumulative :  True면 누적그래프로 그림
*
histtype : 막대그래프 타입(bar, barstacked, step, stepfilled)
*   align : 각 막대 중앙의 위치(left, mid, right)
- orientation : 막대의 방향(vertical,horizontal), 기본값은 vertical, horizontal으로 하면 가로타입 막대그래프로 그려짐
- rwith : 막대의 너비
- log : True면 히스토그램의 축이 로그 스케일(log scale)로 결정
- color : 막대의 색 결정
- label : 각 데이터에 대한 라벨을 붙임. 문자열이고 2개이상일때는 리스트로 넣어야함
- stacked : True일때 다수의 데이터를 겹쳐서 표현할 수 있음.
